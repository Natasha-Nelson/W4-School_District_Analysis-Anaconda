```python
#load pandas
import pandas as pd

# Files to load
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

#read school data
school_data_df = pd.read_csv(school_data_to_load)

#read student data
student_data_df = pd.read_csv(student_data_to_load)

school_data_df.dtypes
```

```python
student_data_df.dtypes
```

```python
#create a list of prefixes and suffixes to remove
prefixes_suffixes = ["Dr. ","Miss ","Mr. ","Mrs. ","Ms. "," DDS"," DVM", "MD"," PhD"]

# Iterate through the "prefixes_suffixes" list and replace them with an empty space, "" when it appears in the student's name.
for word in prefixes_suffixes:
    student_data_df["student_name"] = student_data_df["student_name"].str.replace(word,"")

student_data_df.head(15)
```

```python
# Combine the data into a single dataset.
school_data_complete_df = pd.merge(student_data_df, school_data_df, on=["school_name", "school_name"])
school_data_complete_df.head()
```

```python
# Get the total number of students.
student_count = school_data_complete_df["Student ID"].count()
student_count
```

```python
# Calculate the total number of schools.
school_list = school_data_complete_df["school_name"].unique()
school_count = len(school_list)
school_count
```

```python
# Calculate the total budget.
total_budget = school_data_df["budget"].sum()
total_budget
```

```python
# Calculate the average reading score.
average_reading_score = school_data_complete_df["reading_score"].mean()
average_reading_score
```

```python
# Calculate the average reading score.
average_math_score = school_data_complete_df["math_score"].mean()
average_math_score
```

```python
# Get all the students who are passing math in a new DataFrame.
passing_math = school_data_complete_df[school_data_complete_df["math_score"] >= 70]
passing_math.head()
```

```python
# Get all the students who are passing reading in a new DataFrame.
passing_reading = school_data_complete_df[school_data_complete_df["reading_score"] >= 70]
passing_reading.head()
```

```python
# Calculate the number of students passing math and reading
passing_math_count = passing_math["student_name"].count()
print(passing_math_count)

passing_reading_count = passing_reading["student_name"].count()
print(passing_reading_count)
```

```python
# Calculate the percentage of students passing math and reading
percent_pass_math = passing_math_count/student_count * 100
print(percent_pass_math)

percent_pass_reading = passing_reading_count/student_count * 100
print(percent_pass_reading)
```

```python
# Calculate the number and percentage of students of students passing both math and reading
# filter the complete data frame for math and reading scores above 70
passing_math_reading = passing_reading[passing_reading["math_score"] >= 70]

#find the number of student in this double filtered DF
passing_math_reading_count = passing_math_reading["student_name"].count()
print(passing_math_reading_count)

#calculate the percentage of students passing both subjects
percent_pass_math_reading = passing_math_reading_count / student_count * 100
print(percent_pass_math_reading)
```

```python
# Adding a list of values with keys to create a new DataFrame.
district_summary_df = pd.DataFrame(
          [{"Total Schools": school_count,
          "Total Students": student_count,
          "Total Budget": total_budget,
          "Average Math Score": average_math_score,
          "Average Reading Score": average_reading_score,
          "% Passing Math": percent_pass_math,
         "% Passing Reading": percent_pass_reading,
        "% Overall Passing": percent_pass_math_reading}])
district_summary_df
```
